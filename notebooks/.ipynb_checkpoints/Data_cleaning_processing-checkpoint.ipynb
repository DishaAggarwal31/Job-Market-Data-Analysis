### Standardize Column Names
Ensure that column names are consistent across files. For instance, if all files have a column for job titles, make sure the column name is the same across all files. You can rename columns like so:

`````python
indian_city_state_map = {
    'Mumbai': 'Maharashtra',
    'Pune': 'Maharashtra',
    'Nagpur': 'Maharashtra',
    'Bengaluru': 'Karnataka',
    'Banglore' : 'Karnataka',
    'Gurugram' : 'Haryana',
    'Hyderabad': 'Telangana',
    'Chennai': 'Tamil Nadu',
    'Delhi': 'Delhi',
    'Gurgaon': 'Haryana',
    'Noida': 'Uttar Pradesh',
    'Kolkata': 'West Bengal',
    'Ahmedabad': 'Gujarat',
    'Jaipur': 'Rajasthan',
    'Lucknow': 'Uttar Pradesh',
    'Meerut' : 'Uttar Pradesh'
}
`````

`````python
def extract_city_state_from_mapping(location, city_state_map):
    if pd.isna(location):
        return pd.Series([None, None])

    location = location.lower()
    matched_cities = []
    matched_states = []

    for city in city_state_map:
        if city.lower() in location:
            matched_cities.append(city)

    # Optional: return all or just the first - I am returning all and will later split my data set as per these columns
    for city in matched_cities:
        state = city_state_map[city]
        matched_states.append(state)
    
    if matched_cities:
        return pd.Series([matched_cities, matched_states])
        
    # Handle remote/general case
    if 'remote' in location:
        return pd.Series(['Remote', None])
    if 'india' in location:
        return pd.Series(['India', None])

    return pd.Series([None, None])
`````

`````python
monster_df[['city', 'state']] = monster_df['location'].apply(lambda x: extract_city_state_from_mapping(x, indian_city_state_map))
`````
